6. Jika Harga Jual kurang dari 70000 Maka Status Audit Adalah Tidak, Diatas 70000 Status Audit Adalah Ya (skor : 20)



```python
import pandas as pd

df=pd.read_csv('databarang.csv')
df['Diskon'] = df['Harga Beli'] * 0.1 
df['Pajak'] = df['Harga Beli'] * 0.05
df['Harga Jual'] = df['Harga Beli'] - df['Diskon'] + df['Pajak']
df['Harga Jual'] = df['Harga Jual'].astype("int")

if df['Harga Jual'] > 70000 :
    df['Audit':['Ya']]
else :
    df['Audit':['Tidak']]
df
```

7. Buatlah layar keluaran visualisasi pie chart berdasarkan data file excel tingkat_kelulusan.csv (skor : 10)



```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('tingkat_kelulusan.csv')

data = df['Tingkat Kelulusan']
bulan = df['Bulan']

plt.pie(data, labels=bulan, autopct='%1.1f%%')
plt.show()
```

8. Buatlah layar keluaran visualisasi diagram garis berdasarkan data file excel tingkat_kelulusan.csv skor : 10)



```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('tingkat_kelulusan.csv')

data = df['Tingkat Kelulusan']
bulan = df['Bulan']

x = bulan
y = data

plt.title('Data Mahasiswa Lulus')
plt.xlabel('Bulan')
plt.ylabel('Tingkat Kelulusan')
plt.xticks(rotation=90)
plt.plot(x, y)
plt.show()
```
